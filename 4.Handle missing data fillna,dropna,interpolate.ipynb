```python
# convert day [string] into date:
df = pd.read_csv("weather_data1.csv",parse_dates = ["day"])
df
```

```python
# change the index as day:
df.set_index("day",inplace = True)
df
```

```python
# Replace NaN values with 0:
new_df = df.fillna(0)
new_df
```

```python
# cahnge
new_df = df.fillna({
    "temperature":0,
    "windspeed": 0,
    "event": "no_event"
})
new_df
```

```python
# ffill and bfill these both are used for estimated values.
# it defaultly copies values by vertically.
# if we change the axis then it will change.
```

```python
# ffill: forward fill values from forward ones
new_df = df.fillna(method= "ffill")
new_df
```

```python
# bfill: this will fill values by backward.
new_df = df.fillna(method = "bfill")
new_df
```

```python
# change the axis:
new_df = df.fillna(method = "backfill",axis = "columns")
new_df
```

```python
# limit: if we have two NaN values and limit is 1 it will fill values for 1 position:
new_df = df.fillna(method = "ffill",limit = 1)
new_df
```

```python
new_df = df.fillna(method = "backfill",limit = 2)
new_df
```

```python
# interpolate: it returns the values linearly.
# when there are three NaN values then there will be .025 difference 
# interpolate defaultly will be linear:
new_df = df.interpolate()
new_df
```
