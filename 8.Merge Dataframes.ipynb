```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    "city"        : ['new york','paris','virginia'],
    "temperature" : [21,14,18]
})
df1
```

```python
df2 = pd.DataFrame({
    "city"     : ['new york','paris','virginia'],
    "humidity" : [65,68,75]
})
df2
```

```python
# on = 'city' we use this because we merge using city.
df3 = pd.merge(df1,df2,on = 'city')
df3

# this is exactly join operation in MySql
```

```python
# Case: what if two Data Frames have different cities and we merge them using cities:
# this is similar to inner join[sql] or intersection in set theory
df1 = pd.DataFrame({
    "city"        : ['new york','paris','virginia'],
    "temperature" : [21,14,18]
})
df1

df2 = pd.DataFrame({
    "city"     : ['new york','paris','texas','alaska'],
    "humidity" : [65,68,75,83]
})
df2

df3 = pd.merge(df1,df2,on = 'city')
df3
```

```python
# join DataFrames and make it as outer join[sql] or union in set theory:
df3 = pd.merge(df1,df2,on = 'city',how = "outer")
df3
```

```python
# left and right will be decided by order of data frames u give for merge.
```

```python
# left outer join:
df3 = pd.merge(df1,df2,on = 'city',how = "left")
df3
```

```python
# Right outer join:
df3 = pd.merge(df1,df2,on = 'city',how = "right")
df3
```

```python
#indicate each of the city either it is left or Right.
df3 = pd.merge(df1,df2,on = 'city',how = "outer",indicator = True)
df3
```

```python

df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3 = pd.merge(df1,df2,on = 'city',suffixes = ['_left','_right'])
df3
```

```python
df3 = pd.merge(df1,df2,on = 'city',how = 'outer',suffixes = ['_left','_right'])
df3
```
