```python
# converters these we use to convert something into meaningfully:
```

```python
# this is not good way to do because if there are any companies like RIL is 
# n.a. this will also return the name as sam walton.

def conv_people(cell):
    if cell == "n.a." or cell == "not available":
        return "sam walton"
    return cell
def convert_eps(cell):
    if cell == "not available" or cell == "n.a.":
        return None 
    return cell
df = pd.read_excel("stock_data.xlsx","Sheet1",converters = {
     "people" : conv_people,
     "eps": convert_eps})
df
```

```python
# write these two data frames into same excel file as 2 different sheets
df_stocks = pd.DataFrame({
        'tickers': ['GOOGL', 'WMT', 'MSFT'],
        'price': [845, 65, 64 ],
        'pe': [30.37, 14.26, 30.97],
        'eps': [27.82, 4.61, 2.12]  })
df_weather =  pd.DataFrame({
      'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow'] })
```

```python
# excel writer: we can write with the help of excel writer:
with pd.ExcelWriter("Stocks_Weather.xlsx") as writer:
    df_stocks.to_excel(writer,sheet_name ="Stocks")
    df_weather.to_excel(writer,sheet_name = "Weather")
```
