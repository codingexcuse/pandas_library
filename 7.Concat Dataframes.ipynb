```python
import pandas as pd
```

```python
india_weather = pd.DataFrame ({
    "city"        : ["mumbai","delhi","Banglore"],
    "temperature" : [32,45,30],
    "humidity"    : [80,60,78]
})
india_weather
```

```python
us_weather = pd.DataFrame({
    "city"        : ["new york","chicago","paris"],
    "temperature" : [21,14,35],
    "humidity"    : [68,65,75]
})
us_weather
```

```python
# concantate both DataFrames
df = pd.concat([india_weather,us_weather])
df
```

```python
# modify index from df:
df = pd.concat([india_weather,us_weather],ignore_index = True)
df
```

```python
# display weather according to country:
df = pd.concat([india_weather,us_weather],keys = ["india","usa"])
df
```

```python
temp_df = pd.DataFrame({
    "city"        : ["mumbai","delhi","Banglore"],
    "temperature" : [32,45,30] },index = [0,1,2])
temp_df
```

```python
wind_df = pd.DataFrame({
    "city"        : ["delhi","Banglore"],
    "wind speed"  : [6,12]
},index = [1,2])
wind_df
```

```python
df = pd.concat([temp_df,wind_df])
df
```

```python
# observe above ans this:
df = pd.concat([temp_df,wind_df],axis = 1)
df

# this works perfectly without NaN values when there is correct order in city
```

```python
df = pd.concat([temp_df,wind_df],axis = 1)
df
```
